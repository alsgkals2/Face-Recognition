```python
path = './videos'
temp_list3 = set()
for a,b,c in os.walk(path):
  for _c in c:
    if 'sj' in _c:
      temp_list3.add(os.path.join(a,_c))
            
temp_list3
```

## Source code

```python
import tqdm
save_path = os.path.abspath('.') + '/dataset_with_background/'
os.makedirs(save_path,exist_ok=True)

for _mp in tqdm.tqdm(temp_list3):
  vidcap = cv2.VideoCapture(_mp)
  ret = True
  count = 0
  while(ret):
      ret, image = vidcap.read() # 이미지 사이즈 960x540으로 변경
      if not ret: break
      if(int(vidcap.get(1)) % 10 == 0):
          name = _mp.split('.')[-2].split('/')[-1]
          cv2.imwrite(save_path+f"/{name}_{count}.png", image) #print('Saved frame%d.jpg' % count)
          count += 1

  # print("END")
  cv2.destroyAllWindows()
```
